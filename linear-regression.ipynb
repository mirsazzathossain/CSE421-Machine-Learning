## **Problem Statement**

In this notebook, we will predict sales revenue for a given advertising budget using a simple linear regression model. For this purpose, we will use the kaggle dataset [Advertising.csv](https://www.kaggle.com/ashydv/advertising-dataset). The dataset consists of 200 rows of data, each representing a market with the following attributes:

- `TV`: advertising dollars spent on TV for a single product in a given market (in thousands of dollars)
- `Radio`: advertising dollars spent on Radio
- `Newspaper`: advertising dollars spent on Newspaper
- `Sales`: sales of a single product in a given market (in thousands of widgets)

We will try to predict sales using the three other features.

## **Import necessary packages**

We will use the following packages for this tutorial:

- `pandas` to load and manipulate data
- `numpy` to work with arrays of data
- `matplotlib` and `seaborn` for plotting and visualization
- `sklearn` for machine learning and data analysis tools

```python
data = pd.read_csv('./datasets/Advertising.csv')
data.head()
```

## **Model building**

We will build a simple linear regression model to predict the sales revenue using the advertising budget for TV. In linear regression, we try to fit a line to the data. The line is given by the equation:

$$y=c+m\times X$$

where $y$ is the target variable (sales revenue), $X$ is the feature (advertising budget for TV), $m$ is the slope of the line (coefficient or model parameter), and $c$ is the intercept of the line (model parameter).

## **Prepare the data**

Before we can train a machine learning model, we need to prepare the data. For our model, we need the following two variables:

- `X`: the values of the feature variable (advertising budget for TV)
- `y`: the values of the target variable (sales revenue)

From the dataframe (`data` in our case), we will take the values of the `TV` column as `X` and the values of the `Sales` column as `y`.

```python
X = data['TV']
y = data['Sales']
```

## **Train-Test split**

Before we train the model, we need to split the data into training and testing sets. We will use the `train_test_split()` function from the `sklearn.model_selection` module to split the data. The function takes the following arguments:

- `X`: the values of the feature variable
- `y`: the values of the target variable
- `test_size`: the size of the testing set (0.2 means 20% of the data will be used for testing)
- `random_state`: the seed value for the random number generator used by the function

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## **Fit the model**

We will use the `LinearRegression()` function from the `sklearn.linear_model` module to create a linear regression model. Then we will use the `fit()` function to train the model using the training data.

```python
reg_model = LinearRegression()
reg_model.fit(X_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1))

print(f"c = {reg_model.intercept_}")
print(f"m = {reg_model.coef_}")
```

We can see that the model has a coefficient of $0.05$ and an intercept of $6.77$. So, the equation of the line is:

$$y=6.77+0.05\times X$$

So, we got our model parameters $m$ and $c$ from the training data. Now, we can use the model to predict the sales revenue for a given advertising budget for TV.

## **Predict the sales revenue**

We will use the `predict()` function to predict the sales revenue for the testing data. The function takes the values of the feature variable as the argument and returns the predicted values of the target variable.

```python
y_pred = reg_model.predict(X_test.values.reshape(-1, 1))
```

## **Check the output**

We will plot the testing data and the predicted values of the target variable to check the output of the model.

```python
sns.scatterplot(x=X_test, y=y_test)
sns.lineplot(x=X_test, y=y_pred.reshape(-1), color='r')
plt.show()
```

## **Evaluate the model**

#### **$R^2$ score**
We will use the `score()` function to evaluate the model. The function takes the testing data as the argument and returns the coefficient of determination ($R^2$) of the prediction. The coefficient of determination is a measure of how well the model is performing. The value of $R^2$ is between 0 and 1. The closer the value is to 1, the better the model is performing.

```python
r2 = reg_model.score(X_test.values.reshape(-1, 1), y_test.values.reshape(-1, 1))

print(f"R2 = {r2}")
```

#### **Mean squared error (MSE) & Root mean squared error (RMSE)**
We will use the `mean_squared_error()` function from the `sklearn.metrics` module to calculate the mean squared error (MSE) of the model. The function takes the actual values of the target variable and the predicted values of the target variable as the arguments and returns the MSE. The MSE is the average of the squared difference between the actual and predicted values of the target variable. The RMSE is the square root of the MSE.

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"MSE = {mse}")
print(f"RMSE = {rmse}")
```

#### **Mean absolute error (MAE)**

We will use the `mean_absolute_error()` function from the `sklearn.metrics` module to calculate the mean absolute error (MAE) of the model. The function takes the actual values of the target variable and the predicted values of the target variable as the arguments and returns the MAE. The MAE is the average of the absolute difference between the actual and predicted values of the target variable.

```python
mae = mean_absolute_error(y_test, y_pred)

print(f"MAE = {mae}")
```
